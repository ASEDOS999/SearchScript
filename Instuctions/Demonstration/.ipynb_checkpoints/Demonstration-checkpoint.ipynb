# Extracting Instruction

It demonstrates a workability of our approach to extract action-instructions from text. For each test document there is a set of names of extracted action and for each action there is a sentence involving the action.

All documents was taken from [this google-disk](https://drive.google.com/drive/folders/1JUv5yHlHYR6hHf3d6MhVKiFjo5Zc4hEN). There are original documents names and links to them below.

```python
import sys
sys.path.append('../..')

from TextProcessing import graph_construct_title as GC
from TextProcessing import DFS
from action import construct_sentence as CS

def show_actions(actions, with_print = False):
    text = []
    N = -1
    for act in actions:
        if N >= 0 and text[N][0] == act.section:
            text[N].append({'Name action':act.name_action, 'Sentence':CS(act.sentence)})
        else:
            N += 1
            text.append([act.section])
            text[N].append({'Name action':act.name_action, 'Sentence':CS(act.sentence)})
    return text
```

## 1. File 'Инструкция по покупке подержанного авто для «продвинутого чайника».txt'

The original is [Инструкция по покупке подержанного авто для «продвинутого чайника»](https://docs.google.com/document/d/1AkoVHz2eJh4xmADGxkzjmS4slSN-Lbpd8xPwfjLBcrE/edit).

```python
handle = open("Инструкция по покупке подержанного авто для «продвинутого чайника».txt", "r")
text = handle.read()
handle.close()

root = GC(text).get_graph()
```

```python
list_ = DFS(root)
```

```python
text = show_actions(list_)
for i in text:
    print(i[0])
    print()
    for j in i[1:]:
        print('Name action:', j['Name action'])
        print('Sentence:', j['Sentence'],'\n')
    print('\n')
```
