### 2.2. Script made by hand

Сценарий покупателя:
1. Поиск интересующего авто по объявлениям. О чем важно помнить:
   1. не оставлять задатки продавцу +
   2. оценить среднюю рыночную стоимость автомобиля +
   3. проверить документы (если машина на иностранных номерах - придется потратиться на растаможку) +
   4. если машина на номерах другого региона +
      1. поискать объявления с этими же фотографиями но другими параметрами или владельцами
      2. задавать вопросы продавцу и анализировать ответы
   1. не рассматривать объявление без фото или с плохими фотографиями -
   2. не рассматривать автомобили без документов (даже на запчатси) -
   3. проверить, что нет запрета на регистрационные действия -
   4. проверить, что машина не в залоге/кредите/лизинге +
   5. пробить машину по базе ГИБДД (по VIN и госномеру) +
   6. проверить количество владельцев -
1. Разговор с продавцом
   1. проверить, является ли продавец собственником +
   2. проверить паспорт владельца на сайте ФМС +
   3. переспросить год выпуска и пробег автомобиля +
   4. узнать, в каком регионе эксплуатировалась машина, проверить штрафы -
   5. узнать, что идет в комплекте с машиной +
   6. попросить у владельца фото/скан ПТС, СТС, номерных знаков, или попросить продиктовать номер машины, СТС, VIN, ФИО и дату рожденя владельца -
1. Личная встреча с продавцом
   1. проверить документы: +
      1. ПТС
         1. все номера и данные автомобиля из ПТС совпадают с тем, что указано в объявлении +
         2. посмотреть раздел “Особые отметки” +
         3. посмотреть раздел “Таможенные ограничения” +
         4. проверить отметку “Дубликат” +
         5. ПТС выдан в Бурятии или Чечне +
      1. СТС +
      2. Страховка +
   1. сверить данные из документов с данными на автомобиле
      1. проверить, что на номерах кузова/двигателя/шасси нет следов попытки изменить номер или отдельные символы в нем +
   1. внешний осмотр
      1. поискать следы ремонта: +
         1. разное расстояние между деталями +
         2. детали отличаются по цвету +
         3. отличаются сварные точки в разных проемах +
         4. качество сварных швов разное +
         5. на креплениях есть следы воздействия +
         6. новая краска на старой машине +
         7. коррозия под сидениями/обшивкой +
      1. осмотр салона
         1. оценить возраст по состоянию салона +
         2. проверить работу всех деталей (свет, регулировка сидений, подогрев…) +
1. Оформление автомобиля на нового владельца
   1. составление ДКП:
      1. ДКП составить в трех экземплярах -
      2. не оформлять по генеральной доверенности +
      3. в ДКП и ПТС расписывается только собственник или его законный представитель +
      4. указать в ДКП реальную стоимость авто +
      5. подробно расписать все параметры автомобиля в ДКП (пробег, состояние, ограничения) +
      6. составить акт приема-передачи автомобиля +
      7. оформить расписку о получении денег -
   1. пройти проверку в ГИБДД -
      1. проверка автомобиля сотрудниками ГИБДД +
      2. составление заявления на оформление автомобиля +
      3. предоставить ПТС, паспорт и договор купли-продажи +
      4. получить СТС на свое имя +
