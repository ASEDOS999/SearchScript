# Demonstration of Extracting

```python
import sys
sys.path.append('../')
from ScriptExtract.Preprocessing.TextProcessing import table
```

## Text №1

```python
handle = open("../Texts/text2_2.txt", "r")
text = handle.read()
print(len(text))
handle.close()
a = table().extract_one(text)
t = a[0]
a[1]
```

```python
k = 0
s = 0
for i in t:
    if len(i['Actions']):
        k += 1
        print("%d. "%(k)+i['Sentence'])
        for act in i['Actions']:
            b, e = act.inform['VERB'][0].begin, act.inform['VERB'][0].end
            verb = i['Sentence'][b:e]
            print(verb)
            print(act.type_action)
            for j in i['Relations']:
                if verb == i['Sentence'][j['parent']['start']:j['parent']['end']]:
                    print('Type: %s; Parent: %s; Child: %s'%(j['tp'], 
                                                             i['Sentence'][j['parent']['start']:j['parent']['end']],
                                                             i['Sentence'][j['child']['start']:j['child']['end']]))                

        s += len(i['Sentence'])
        print()
```
