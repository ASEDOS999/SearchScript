# Demonstration of Extracting

```python
import sys
import numpy as np
sys.path.append('../')
from ScriptExtract.Preprocessing.TextProcessing import table
```

## Text №1

```python
handle = open("../Texts/text2_2.txt", "r")
text = handle.read()
print(len(text))
handle.close()
a = table(use_sem = False).extract_one(text)
t = a[0]
a[1]
```

```python
k = 0
s = 0
for i in t:
    if len(i['Actions']):
        k += 1
        print("%d. "%(k)+i['Sentence'])
        for act in i['Actions']:
            b, e = act.inform['VERB'][0].begin, act.inform['VERB'][0].end
            verb = i['Sentence'][b:e]
            print('\n', verb)
            print(act.type_action)
            for depend in act.inform:
                if not depend in ['punct', 'VERB']:
                    print('\n' + depend + ':')
                    for w in act.inform[depend]:
                        print('Word Lemma:', w[0].lemma)
                        print('Full dependence:', ' '.join(list(np.array(act.sentence)[w[1]])))
        s += len(i['Sentence'])
        print('\n')
```
