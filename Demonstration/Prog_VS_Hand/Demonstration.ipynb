```python
import sys
sys.path.append('../../Parser')
from some_reparser import extract_semantic_relations
```

```python
k = 0
for i in a[0]:
    if i['IsInstr']:
        k += 1
        print("%d. "%(k)+i['Sentence'])
        sent = i['Sentence']
        relations = extract_semantic_relations(sent)
        for relation in relations:
            print('TP:',relation['tp'])
            print('Parent:',[sent[relation['parent']['start']:relation['parent']['end']]])
            print('Child:',[sent[relation['child']['start']:relation['child']['end']]])
            print()
        print()
```
