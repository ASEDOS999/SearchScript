# Demonstration of Extracting

```python
import sys
sys.path.append('../../')
from ScriptExtract.Preprocessing.TextProcessing import table
```

## Text №1

```python
handle = open("../../Texts/text2_2.txt", "r")
text = handle.read()
print(len(text))
handle.close()
a = table().extract_one(text)
t = a[0]
a[1]
```

```python
k = 0
s = 0
for i in t:
    if i['FirstLevel']:
        k += 1
        print("%d. "%(k)+i['Sentence'])
        for j in i['Relations']:
            print('Type: %s; Parent: %s; Child: %s'%(j['tp'], 
                                                     i['Sentence'][j['parent']['start']:j['parent']['end']],
                                                    i['Sentence'][j['child']['start']:j['child']['end']]))
        s += len(i['Sentence'])
        print()
```

```python
k = 0
for i in t:
    if i['SecondLevel']:
        k += 1
        print("%d. "%(k)+i['Sentence'])
        for j in i['Relations']:
            print('Type: %s; Parent: %s; Child: %s'%(j['tp'], 
                                                     i['Sentence'][j['parent']['start']:j['parent']['end']],
                                                    i['Sentence'][j['child']['start']:j['child']['end']]))
        s+= len(i['Sentence'])
        print()
s/len(text)
```

```python
prev, next_ = None, None
for ind,j in enumerate(t):
    if j['FirstLevel']:
        next_ = k
        break
for k,i in enumerate(t):
    if i['FirstLevel']:
        prev = k
        next_ = None
        for ind,j in enumerate(t[k+1:]):
            if j['FirstLevel']:
                next_ = ind+k+1
                break
    elif i['SecondLevel'] and not i['FirstLevel']:
        print(k,i['Sentence'])
        if not next_ is None:
            print('NEXT(%d):'%(next_-k), t[next_]['Sentence'])
        if not prev is None:
            print('PREV(%d):'%(k-prev), t[prev]['Sentence'])
        print()
```
